```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df = pd.read_csv('DisneylandReviews.csv', encoding='latin-1')
df.head()
```

```python
df[["Reviewer_Location", 'Review_Text', 'Rating']].head()
```

```python
DisneyLand = df.groupby("Reviewer_Location")
DisneyLand.describe().head()
```

```python
DisneyLand.mean().sort_values(by="Rating", ascending=False)
DisneyLand.head()
```

```python
text = df.Review_Text[0]

wordcloud = WordCloud().generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
